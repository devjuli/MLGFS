# Das Gadientenverfahren in zwei Dimensionen:

```python
import numpy as np
from matplotlib import pylab
%matplotlib inline
```

```python
def f(x):
    return 0.5*x**2

x = np.linspace(-10,10,100)


x_prime = -10
alpha = 0.1

pylab.plot(x,f(x))
for i in range(20):
    deltax = alpha*x_prime
    pylab.arrow(x_prime,f(x_prime),-deltax,f(x_prime)-f(x_prime+deltax),linewidth=4)
    x_prime -= deltax
pylab.show()
```

# Das Gradientenverfahren in drei Dimensionen

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


def f(x, y):
    return 0.5*x**2+0.5*y**2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax = plt.axes(projection="3d")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')

#Gradientenverfahren:
x = -5
y = 5
alpha = 0.1

for i in range(10): 
    x -= alpha * x
    y -= alpha * y
    ax.scatter(x,y,f(x,y))
plt.show()
```

# Visualisierung des Gewichtungsvektors

```python
from sklearn.datasets import load_digits
import numpy as np
from matplotlib import pylab
%matplotlib inline
digits = load_digits()
xdata = digits.data[:100]
target_digit = 0

def sigmoid(x):
    return 1/(1+np.exp(-x))

def fit_and_show(x,y,n_iter=1000,alpha=1e-1):
    b = np.random.random()*2-1
    w = np.random.random((x.shape[1]))*2-1
    for i in range(n_iter):
        z = x.dot(w)+b
        sig_z = sigmoid(z)
        errorgrad = -(y-sig_z)
        w -= errorgrad.dot(x)*alpha
        b -= errorgrad*alpha
    pylab.gray()
    pylab.imshow(w.reshape([8,8]))
    pylab.show()
```

```python
for target_digit in range(10):
    ydata = np.where(digits.target[:100] == target_digit,1,0)
    fit_and_show(xdata,ydata)
```
